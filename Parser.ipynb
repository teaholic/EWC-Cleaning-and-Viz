```python
def collider_no(data):
    numbers = []
    numbers.append([number for number in data if len(re.findall(r'[0-9+\s]', number)) == len(number) and len(number) != 0])
    return numbers

def collider_abc(data):
    abc = []
    abc.append([word for word in data if len(re.findall(r'[a-z]', word)) == len(word)])
    return abc
```

```python
numbers = map(collider_no, parsed)
num = [x for section in numbers for item in section for x in item]
def clean_nums(num): 
    return (lambda code: code[:], num)[1].rstrip().lstrip()
num = map(clean_nums, num)
num = [x for x in num if len(x) != 0]
    
number = []    
for x in num:
    if len(x) < 9 : number.append(x)
    else:
        a = x[:7]
        b = x[9:]
        number.append(a)
        number.append(b)
number
```

```python
def split_nums(num):
    number = []
    if len(num) < 9 : number.append(num)
    else:
        a = num[:7]
        b = num[8:]
        number.append(a)
        number.append(b)
    return number
```

```python
def prep_nums(num):
    clean = clean_nums(num)
    return split_nums(clean)
```

```python
def prep_words(words):
    clean = []
    for word in words:
        if len(re.findall(r'[a-z]', word)) == len(word): 
            clean.append(word)
        else: 
            num = clean_nums(word)
            if len(num) < 9: 
                clean.append(num)
            elif len(num) < 11:
                a = num[:5]
                b = num[6:]
                clean.append(a)
                clean.append(b)
            elif len(num) < 15:
                a = num[:4]
                b = num[5:]
                clean.append(a)
                clean.append(b)
            else:
                a = num[:8]
                b = num[9:]
                clean.append(a)
                clean.append(b)
    return clean   
```
